```python
import random

def split_students_randomly(students, num_groups):
    random.shuffle(students)  # Shuffle students randomly
    group_size = len(students) // num_groups  # Base size of each group
    remainder = len(students) % num_groups  # Remaining students to distribute

    groups = []
    start = 0
    for i in range(num_groups):
        extra = 1 if i < remainder else 0 
        groups.append(students[start:start + group_size + extra])
        start += group_size + extra

    return groups

# Example usage
# students = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah", "Ian", "Jack"]
num_groups = 6
groups = split_students_randomly(students, num_groups)

for i, group in enumerate(groups, 1):
    print(f"Group {i}: {group}")

```
