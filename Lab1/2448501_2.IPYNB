```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def extract_frames(video_path, interval=30):
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    extracted_frames = []
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        if frame_count % interval == 0:
            extracted_frames.append(frame)
        
        frame_count += 1
    
    cap.release()
    return extracted_frames

def split_and_display(frame):
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    red_channel = frame_rgb.copy()
    green_channel = frame_rgb.copy()
    blue_channel = frame_rgb.copy()
    
    red_channel[:, :, 1:] = 0  # Keep only the red channel
    green_channel[:, :, [0, 2]] = 0  # Keep only the green channel
    blue_channel[:, :, :2] = 0  # Keep only the blue channel
    
    fig, axs = plt.subplots(1, 4, figsize=(12, 4))
    axs[0].imshow(frame_rgb)
    axs[0].set_title("Original Frame")
    axs[1].imshow(red_channel)
    axs[1].set_title("Red Channel")
    axs[2].imshow(green_channel)
    axs[2].set_title("Green Channel")
    axs[3].imshow(blue_channel)
    axs[3].set_title("Blue Channel")
    
    for ax in axs:
        ax.axis("off")
    
    plt.show()

video_path = r"D:\MSAIM\trimister-3 msaiml\Computer_vision\Computer_vision_practical\video\dummy_tree.mp4"  
frames = extract_frames(video_path, interval=60)  
if frames:
    split_and_display(frames[0])  
```

```python
if frames:
    for frame in frames:
        split_and_display(frame)  
```
